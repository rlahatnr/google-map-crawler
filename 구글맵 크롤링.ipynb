```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
import random
from tqdm.auto import tqdm
from bs4 import BeautifulSoup

file_name= input('가져올 파일 이름 입력하세요: ')
result_name= input('내보낼 파일 이름 입력하세요: ')

# 주석 단축키는 ctrl+/ 입니다!
# input값으로 계속 넣기 번거롭다면 9열을 주석처리한 후 아래처럼 고정해두셔도 됩니다
# file_name= '예시 파일 이름'

# 컴퓨터 환경에 따라 숫자를 조절하세요 (각 페이지 넘길 때마다 로딩 기다리는 시간입니다)
interval = 1

driver = webdriver.Chrome()
driver.get('https://www.google.com/maps/')
driver.implicitly_wait(30)
time.sleep(random.random()+interval)
search=driver.find_element_by_id('searchboxinput')

# .csv
# keyword = pd.read_csv(f'{file_name}.csv', engine='python').name[ 최종인덱스값+1 (예: 최종인덱스 25일 경우 26 입력):]
keyword = pd.read_csv(f'{file_name}.csv', engine='python', encoding='utf-8-sig').name

# .xlsx
# keyword = pd.read_excel(f'{file_name}.xlsx').name

box=[]
for key in tqdm(keyword):
    #검색
    search.clear()
    search.send_keys(key)
    search.send_keys(Keys.ENTER)
    driver.implicitly_wait(30)
    time.sleep(random.random()+2)
    source = driver.page_source
    html = BeautifulSoup(source, "lxml")
    
    #다중결과 일 때
    try:
        if html.select('h1.x3AX1-LfntMc-header-title-title.gm2-headline-5') == []:
            lists = driver.find_elements_by_class_name('CUwbzc-content.gm2-body-2')
            if len(lists) == 0:
                box.append([key, '-', '-'])
            for unit in lists:
                try:
                    name= unit.find_elements_by_class_name('qBF1Pd.gm2-subtitle-alt-1')[0].text
                    juso= unit.find_elements_by_class_name('ZY2y6b-RWgCYc')[2].text
                    if '·' in juso:
                        juso=juso.split('·')[1].strip(' ')
                    elif '호텔' in juso:
                        continue
                    box.append([key, name, juso])
                except Exception as e:
                    pass

        #단일결과 일 때
        else:
            juso = driver.find_element_by_class_name('QSFF4-text.gm2-body-2').text
            box.append([key, driver.find_element_by_class_name('x3AX1-LfntMc-header-title-title.gm2-headline-5').text, juso])
    except Exception as e:
        box.append([key, '-', '-'])
        pass
    pd.DataFrame(box, columns=['검색어', '검색된 값', '주소']).to_excel(f'{result_name}.xlsx')
```
